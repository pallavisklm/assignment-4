```python
import matplotlib.pyplot as plt
import numpy as np
departments=["cse","civil","ece","mech"]
students=[20,49,25,25]
plt.bar(departments,students,color="pink")
plt.title("rgukt")
plt.xlabel("departments")
plt.ylabel("students")
plt.show()
```

```python
plt.bar(departments,students,color="red",width=0.3)
plt.show()
```

```python
plt.bar(departments,students,width=0.3,color=['g','b','r','y'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
departments=["cse","civil","ece","mech"]
students=[20,49,25,25]
plt.barh(departments,students,color="pink")
plt.title("rgukt")
plt.xlabel("departments")
plt.ylabel("students")
plt.show()
```

```python
a=[5,7,8,9,5]
b=[3,6,7,9,4]
x=np.linspace(1,50,5)
plt.bar(x,a,width=4,color="c",label="austrailia")
plt.bar(x+3,b,width=4,color="g",label="india")
plt.legend()
plt.show()
```
